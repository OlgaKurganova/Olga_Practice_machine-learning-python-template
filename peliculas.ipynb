```python
import numpy as np
import pandas as pd

dataframe=pd.read_csv("Netflix movies data.csv")
print(dataframe)

```

```python
#2.2 Filter the data into TV Shows and store it as netflix_tv and created another dataframe for Movies and store as netflix_subset. 

#df1=pd.DataFrame(dataframe,columns=['show_id'])
#df1.head()
netflix_tv=dataframe[dataframe['type']=='TV Show']
netflix_tv.head()
netflix_subset=dataframe[dataframe['type']=='Movie']
netflix_tv.head()
```

```python
#3.3Calculate the Various Statistical Characteristics of the Duration of Movies and the for TV Shows. Write your observations.
from scipy.stats import skew
import statistics as stats


duration_movies=netflix_tv['duration']
duration_show=netflix_subset['duration']
duration_movies.head()
duration_show.head()

print('Media', np.mean(duration_movies))
print('Desviacion estandart',np.std(duration_movies))
print('Minimo', np.min(duration_movies))
print('Maximo', np.max(duration_movies))
print('Rango', np.ptp(duration_movies))
print('Mode', stats.mode(duration_movies))


print('Media', np.mean(duration_show))
print('Desviacion estandart', np.std(duration_show))
print('Minimo', np.min(duration_show))
print('Maximo',np.max(duration_show))
print('Rango',np.ptp(duration_show))
print('Mode', stats.mode(duration_show))
```
